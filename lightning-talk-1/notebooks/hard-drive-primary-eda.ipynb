# Hard Drive Failure Prediction
---

## Problem Statement

**Problem:** Hard drive failures cause catastrophic data loss and significant downtime for businesses. Detecting imminent failures before they happen enables preventive replacement, protecting data and reducing operational costs. The S.M.A.R.T. (Self-Monitoring, Analysis, and Reporting Technology) system embedded in hard drives continuously reports diagnostic attributes.

**Question:** Can we use S.M.A.R.T. diagnostic attributes to build a machine learning model that **predicts whether a hard drive will fail** — enabling data center operators to proactively replace at-risk drives before failure occurs?

**Goal:** Develop a **binary classification** model that predicts hard drive failure (`failure = 1`) vs. normal operation (`failure = 0`) using S.M.A.R.T. sensor readings.

**Approach:** Supervised Machine Learning — Binary Classification on heavily imbalanced data (failures are rare events). Techniques: SMOTE, class weighting, threshold tuning, ensemble models (Random Forest, XGBoost).

---

**Data Source:** [Kaggle — Backblaze Hard Drive Test Data](https://www.kaggle.com/datasets/backblaze/hard-drive-test-data)

```python
# Distribution of top SMART predictors by failure status
known_predictors = ['smart_5_raw', 'smart_187_raw', 'smart_197_raw', 'smart_198_raw', 'smart_9_raw']
plot_cols = [c for c in known_predictors if c in df.columns]

fig, axes = plt.subplots(1, len(plot_cols), figsize=(4 * len(plot_cols), 5))
if len(plot_cols) == 1:
    axes = [axes]

for i, col in enumerate(plot_cols):
    plot_data = df[[col, target]].copy()
    plot_data = plot_data[plot_data[col] < plot_data[col].quantile(0.99)]  # clip extreme outliers for viz
    for val, label, color in [(0, 'Operational', '#2ecc71'), (1, 'Failed', '#e74c3c')]:
        subset = plot_data[plot_data[target] == val][col]
        axes[i].hist(subset, bins=30, alpha=0.6, label=label, color=color, density=True)
    axes[i].set_title(col.replace('smart_', 'SMART ').replace('_raw', ' Raw'), fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].legend()

plt.suptitle('Distribution of Key SMART Attributes by Failure Status', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

### Are there outliers? How would you handle them?

```python
# Focus on raw SMART features
raw_available = [c for c in raw_cols if c in df.columns]

outlier_summary = []
for col in raw_available:
    col_data = df[col].dropna()
    q1  = col_data.quantile(0.25)
    q3  = col_data.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    n_out = ((col_data < lower) | (col_data > upper)).sum()
    pct   = (n_out / len(col_data)) * 100
    outlier_summary.append({'Feature': col, 'Outlier Count': n_out, 'Outlier %': round(pct, 2)})

outlier_df     = pd.DataFrame(outlier_summary).sort_values('Outlier %', ascending=False)
outlier_present = outlier_df[outlier_df['Outlier Count'] > 0]

print(f'SMART raw features with outliers: {len(outlier_present)} / {len(raw_available)}')
print()
print(outlier_present.head(20).to_string(index=False))
```

```python
# Boxplots for most critical SMART attributes
critical_cols = ['smart_5_raw', 'smart_187_raw', 'smart_197_raw',
                 'smart_198_raw', 'smart_9_raw', 'smart_194_raw']
plot_cols     = [c for c in critical_cols if c in df.columns]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes      = axes.flatten()

for i, col in enumerate(plot_cols):
    # Clip for better visualization
    clip_val = df[col].quantile(0.99)
    plot_data = df[[col, target]].copy()
    plot_data[col] = plot_data[col].clip(upper=clip_val)
    sns.boxplot(x=plot_data[target], y=plot_data[col],
                ax=axes[i], palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(col.replace('smart_', 'SMART ').replace('_raw', ' Raw'), fontweight='bold')
    axes[i].set_xlabel('failure (0=OK, 1=Failed)')

plt.suptitle('Critical SMART Attribute Distributions by Failure Status\n(clipped at 99th pct for visualization)',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n--- Outlier Handling Plan ---')
print()
print('    IMPORTANT: For hard drive failure prediction, outliers in SMART attributes')
print('    are often the SIGNAL, not noise — extremely high error counts indicate failure.')
print()
print('Recommended approach by attribute type:')
print()
print('  smart_5_raw (Reallocated Sectors)   → DO NOT remove — high values = failure signal')
print('  smart_187_raw (Uncorrectable Errors) → DO NOT remove — same reason')
print('  smart_197_raw (Pending Sectors)      → DO NOT remove — same reason')
print('  smart_198_raw (Offline Uncorrectable) → DO NOT remove — same reason')
print('  smart_9_raw (Power-On Hours)         → Cap via Winsorization (very old drives)')
print('  smart_194_raw (Temperature)          → Cap at physically impossible values (>80°C)')
print()
print('General strategy: Use RobustScaler (IQR-based) when scaling;')
print('apply log1p transform to right-skewed error count columns.')
```
