## Problem Statement

**Problem:** Phishing attacks are among the most prevalent cybersecurity threats, where malicious actors create fraudulent websites that mimic legitimate ones to steal user credentials and sensitive information. Organizations and individuals lose billions of dollars annually due to phishing.

**Question:** Can we build a machine learning model that accurately classifies a web page as **phishing** or **legitimate** based on its URL and structural features — enabling automated, real-time detection to protect users before they interact with malicious sites?

**Goal:** Develop a **binary classification** model that predicts whether a given web page is phishing (`1`) or legitimate (`0`), using features extracted from URLs and webpage metadata.

**Approach:** Supervised Machine Learning — Binary Classification (e.g., Logistic Regression, Random Forest, Gradient Boosting).

---

**Data Source:** [Kaggle — Web Page Phishing Dataset](https://www.kaggle.com/datasets/danielfernandon/web-page-phishing-dataset)

### What is your target column and what does it represent?

### ❓ Q2 — Is your target categorical or continuous? Is it balanced?

```python
print('The target column "status" is CATEGORICAL (binary: 0 or 1).')
print('This is a CLASSIFICATION problem.\n')

target_counts = df[target].value_counts()
target_pct    = df[target].value_counts(normalize=True) * 100

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Count plot
axes[0].bar(['Legitimate (0)', 'Phishing (1)'], target_counts.values,
            color=['#2ecc71', '#e74c3c'], edgecolor='black', width=0.5)
axes[0].set_title('Target Class Distribution (Count)', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Count')
for i, v in enumerate(target_counts.values):
    axes[0].text(i, v + 50, str(v), ha='center', fontweight='bold')

# Pie chart
axes[1].pie(target_counts.values,
            labels=['Legitimate (0)', 'Phishing (1)'],
            autopct='%1.1f%%',
            colors=['#2ecc71', '#e74c3c'],
            startangle=140, explode=(0.05, 0.05))
axes[1].set_title('Target Class Distribution (%)', fontsize=13, fontweight='bold')

plt.suptitle('Is the Target Balanced?', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print(f'\nClass balance:')
for label, pct in zip(['Legitimate (0)', 'Phishing (1)'], target_pct.values):
    print(f'  {label}: {pct:.1f}%')

ratio = target_counts.max() / target_counts.min()
print(f'\nImbalance ratio: {ratio:.2f}x')
if ratio < 1.5:
    print('➡ The dataset is WELL BALANCED — no special handling needed.')
else:
    print('➡ The dataset is IMBALANCED — consider SMOTE, class_weight, or resampling.')
```

### ❓ Q3 — What types of features do you have? How many of each?

```python
# Drop non-feature columns
feature_df = df.drop(columns=['url', target], errors='ignore')

# Binary columns (only contain 0/1) are treated as categorical
binary_cols     = [c for c in feature_df.columns
                   if feature_df[c].dropna().isin([0, 1]).all()
                   and feature_df[c].nunique() <= 2]
continuous_cols = [c for c in feature_df.columns if c not in binary_cols]

print(f'Total features     : {feature_df.shape[1]}')
print(f'Categorical/Binary : {len(binary_cols)}')
print(f'Numerical/Continuous: {len(continuous_cols)}')
print()
print('--- Categorical / Binary features ---')
print(binary_cols)
print()
print('--- Numerical / Continuous features ---')
print(continuous_cols)

# Bar chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(['Categorical / Binary', 'Numerical / Continuous'],
       [len(binary_cols), len(continuous_cols)],
       color=['#3498db', '#e67e22'], edgecolor='black', width=0.4)
ax.set_title('Feature Types Distribution', fontsize=13, fontweight='bold')
ax.set_ylabel('Number of Features')
for i, v in enumerate([len(binary_cols), len(continuous_cols)]):
    ax.text(i, v + 0.3, str(v), ha='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

### ❓ Q4 — How many null values does each feature have? What's your plan for each?

```python
null_counts = df.isnull().sum()
null_pct    = (null_counts / len(df)) * 100

null_df = pd.DataFrame({
    'Null Count' : null_counts,
    'Null %'     : null_pct.round(2)
}).sort_values('Null Count', ascending=False)

null_present = null_df[null_df['Null Count'] > 0]

if null_present.empty:
    print('✅ No null values found in the dataset! No imputation needed.')
else:
    print(f'⚠ {len(null_present)} features have null values:\n')
    print(null_present)

    fig, ax = plt.subplots(figsize=(12, 5))
    null_present['Null %'].plot(kind='bar', ax=ax, color='#e74c3c', edgecolor='black')
    ax.set_title('Missing Values per Feature (%)', fontsize=13, fontweight='bold')
    ax.set_ylabel('Missing %')
    ax.set_xlabel('Feature')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print('\n--- Null Handling Plan ---')
    for col in null_present.index:
        pct = null_present.loc[col, 'Null %']
        if pct > 50:
            print(f'  {col} ({pct:.1f}%) → DROP column (too many missing values)')
        elif col in binary_cols:
            print(f'  {col} ({pct:.1f}%) → FILL with MODE (binary feature)')
        else:
            print(f'  {col} ({pct:.1f}%) → FILL with MEDIAN (numerical feature)')
```

### ❓ Q5 — Which features are most correlated with the target, and with each other?

```python
# Correlation with target
num_df = df.drop(columns=['url'], errors='ignore').select_dtypes(include=[np.number])
target_corr = num_df.corr()[target].drop(target).sort_values(key=abs, ascending=False)

print('Top 15 features correlated with target (status):')
print(target_corr.head(15).to_string())

fig, ax = plt.subplots(figsize=(10, 7))
colors = ['#e74c3c' if v >= 0 else '#3498db' for v in target_corr.head(15).values]
target_corr.head(15).plot(kind='barh', ax=ax, color=colors, edgecolor='black')
ax.axvline(0, color='black', linewidth=0.8)
ax.set_title('Top 15 Feature Correlations with Target (status)', fontsize=13, fontweight='bold')
ax.set_xlabel('Pearson Correlation Coefficient')
plt.tight_layout()
plt.show()
```

```python
# Heatmap of top correlated features among themselves
top_features = target_corr.head(15).index.tolist() + [target]
corr_matrix  = num_df[top_features].corr()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

fig, ax = plt.subplots(figsize=(14, 10))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f',
            cmap='RdBu_r', center=0, vmin=-1, vmax=1,
            linewidths=0.5, ax=ax)
ax.set_title('Correlation Heatmap — Top Features + Target', fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### ❓ Q6 — Are there outliers? How would you handle them?

```python
# Use IQR method to detect outliers in continuous features
outlier_summary = []
for col in continuous_cols:
    q1  = df[col].quantile(0.25)
    q3  = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    n_outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    pct = (n_outliers / len(df)) * 100
    outlier_summary.append({'Feature': col, 'Outlier Count': n_outliers, 'Outlier %': round(pct, 2)})

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outlier %', ascending=False)
outlier_present = outlier_df[outlier_df['Outlier Count'] > 0]

print(f'Features with outliers: {len(outlier_present)} / {len(continuous_cols)}')
print()
print(outlier_present.head(20).to_string(index=False))
```

```python
# Boxplots for top 6 features with most outliers
top_outlier_cols = outlier_present['Feature'].head(6).tolist()

fig, axes = plt.subplots(2, 3, figsize=(16, 8))
axes = axes.flatten()

for i, col in enumerate(top_outlier_cols):
    sns.boxplot(x=df[target], y=df[col], ax=axes[i],
                palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(f'{col}', fontweight='bold')
    axes[i].set_xlabel('status (0=Legit, 1=Phish)')

plt.suptitle('Boxplots — Top Features with Outliers (by Target Class)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print('\n--- Outlier Handling Plan ---')
for _, row in outlier_present.head(10).iterrows():
    pct = row['Outlier %']
    col = row['Feature']
    if pct > 10:
        print(f'  {col} ({pct:.1f}%) → Keep as-is or use RobustScaler — likely real variance, not errors')
    else:
        print(f'  {col} ({pct:.1f}%) → Cap using IQR Winsorization (clip to 1st–99th percentile)')
```
